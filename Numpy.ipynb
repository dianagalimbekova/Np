```python
import numpy as np
a = np.linspace(0, 2, 201)
phi = np.linspace(-2 * np.pi, 2 * np.pi, 201)
phi = phi[phi != 0]

for k in range(6):
    r = k/4 + np.cos(phi)
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    import matplotlib.pyplot as plt
    plt.plot(x, y)
    plt.show()
```
